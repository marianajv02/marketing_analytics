```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import rpy2.robjects as ro
from rpy2.robjects import pandas2ri
```

# Data conversion from R to Pandas

```python
# Read the Rdata from R data
filepath = './CreditCard_Segments.RData'
ro.r['load'](filepath)

# Accessing the objects training and validation
data = ro.r['CrediCard_Training']
dataNewCust = ro.r['CreditCard_NewCustomers']

# Converting it to pandas

df_train = pandas2ri.rpy2py(data)
df_New = pandas2ri.rpy2py(dataNewCust)
```

Creating dashboards for shiny

```python
if df_train['CREDIT_LIMIT'].isnull().any():
    # Handle missing values as needed
    df_train['CREDIT_LIMIT'].fillna(0, inplace=True)  # Replace NaN values with 0, for example


plt.hist(df_train['CREDIT_LIMIT'], bins=30, edgecolor='black')  # You can adjust the number of bins as needed


plt.xlabel('Credit Limit')
plt.ylabel('Frequency')
plt.title('Histogram of Credit Limit')
plt.show()
```

```python
plt.figure(figsize=(8, 8))  # Set the figure size if needed
plt.pie(df_train['CREDIT_LIMIT'].value_counts(), labels=df_train['CREDIT_LIMIT'].value_counts().index, autopct='%1.1f%%', startangle=90)

# Set plot title
plt.title('Distribution of Credit Limits')

# Display the pie chart
plt.show()
```
